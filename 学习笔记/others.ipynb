### 1. Counter


```python
from collections import Counter

cnt = Counter() 
# Counter({1:2}) 1出现了2次

# 加入元素
cnt[i] += 1

# 判断元素
if i in cnt:
    pass
# 判空
if i not cnt:
    pass
```

### 2. defaultdict

常用于计数，快于 Counter


```python
from collections import defaultdict

d = defaultdict(int)  # 默认所有元素初始化为0

s = "test"

for k in s:
    d[k] += 1

print(d)
```

### 4. 全局变量

python 不可变参数(例如 str,int)作为函数参数传递无法改变全局的值

需要在函数内部和外部声明


```python
def maxDepth(self, root: Optional[TreeNode]) -> int:
    i = 0
    global maxs
    maxs = 0
    self.xian(root, i)
    return maxs


def xian(self, root, i):
    global maxs  # 声明全局变量 函数内部修改全局也会改

    if not root:
        maxs = max(maxs, i)
        return []
    i = i+1
    return [root.val] + self.xian(root.left, i) + self.xian(root.right, i)
```

### 5. List remove 移除问题


```python
def test(g, s):
    gs = sorted(g)
    ss = sorted(s)
    for i in gs:
        for j in ss:
            print(j, i)
            if j < i:
                ss = ss.remove(j)
            else:
                ss = ss.remove(j)
                break


test(g=[7, 8, 9, 10], s=[5, 6, 7, 8])

'''//打印的j,i
5 7
7 7
6 8
8 9
移除5后下一个遍历6 -> 移除5后ss为【6,7,8】 此时应该遍历第二个元素即7 移除7后遍历第下一个元素6

涉及循环遍历时慎用'''


def test(g, s):  # 准确做法
    gs = sorted(g)
    ss = sorted(s)
    for i in gs:
        for j in ss:
            print(j, i)
            if j < i:
                ss = ss[1:]
            else:
                ss = ss[1:]
                break
```

### 7. 计算时间差


```python
from datetime import *
l2 = sorted(l)
sums = 0
for i in range(0, len(l2), 4):
    d1 = datetime.strptime(l2[i], f'%Y-%m-%d %H:%M:%S')  # 转换为datetime
    d2 = datetime.strptime(l2[i+1], f'%Y-%m-%d %H:%M:%S')
    d3 = datetime.strptime(l2[i+2], f'%Y-%m-%d %H:%M:%S')
    d4 = datetime.strptime(l2[i+3], f'%Y-%m-%d %H:%M:%S')
    s1 = d2-d1
    s2 = d4-d3
    sums += s1.seconds + s2.seconds
sums
```

### 10.质数判断


```python
# 遍历2到根号x:
x = 10


def is_prime(x):
    for i in range(2, int(x**0.5)+1):
        if x % i == 0:
            return False
```

### 12.求解一个数的因子问题


```python
import math
num = 10
nums = []
for j in range(1, int(math.sqrt(num))+1):
    if num % j == 0:
        nums.append(j)
        if j * j < num:  # 避免j * j = num重复计数
            nums.append(num//j)
```
